## Model BAsed Forecasting

## Linear Model

```python
train = df1.head(98)
test = df1.tail(24)
```

```python
import statsmodels.formula.api as smf
import patsy
linear_model = smf.ols('Close~t', data = train).fit()#ols = ordinary least squre
linear_pred = linear_model.predict(test['t'])
linear_rmse = np.sqrt(np.mean((np.array(test['Close'])-np.array(linear_pred))**2))
linear_rmse
```

## Exponetial Model

```python
from sklearn.metrics import mean_absolute_percentage_error
import statsmodels.formula.api as smf
```

```python
Exp = smf.ols('log_Close~t', data = train).fit()
pred_Exp = pd.Series(Exp.predict(pd.DataFrame(test['t'])))
rmse_Exp = mean_absolute_percentage_error(test['Close'],pred_Exp)**2
(rmse_Exp)*100
```

## Quadratic 

```python
Quad = smf.ols('Close~t+t_squared', data = train).fit()
pred_Quad = pd.Series(Quad.predict(test[["t", "t_squared"]]))
rmse_Quad = mean_absolute_percentage_error(test['Close'],pred_Quad)**2
(rmse_Quad)*100
```

## Additive Seasonality

```python
add_sea = smf.ols('Close~Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov', data =train).fit()
pred_add_sea = pd.Series(add_sea.predict(test[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov']]))
rmse_add_sea = mean_absolute_percentage_error(test['Close'],pred_add_sea)**2
(rmse_add_sea)*100
```

## Additive Seasonality Quadratic

```python
add_sea_quad = smf.ols('Close~t+t_squared+Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov+Dec',data=train).fit()
pred_add_sea_quad = pd.Series(add_sea_quad.predict(test[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec','t','t_squared']]))
rmse_add_sea_quad = mean_absolute_percentage_error(test['Close'],pred_add_sea_quad)**2
(rmse_add_sea_quad)*100
```

## Multiplicative Seasonality

```python
Mul_sea = smf.ols('log_Close~Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov+Dec',data = train).fit()
pred_Mult_sea = Mul_sea.predict(test)
rmse_Mult_sea = np.sqrt(np.mean((np.array(test['Close'])-np.array(np.exp(pred_Mult_sea)))**2))
rmse_Mult_sea
```

## Multiplicative Additive Seasonality

```python
Mul_Add_sea = smf.ols('log_Close~t+Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov+Dec',data = train).fit()
pred_Mult_add_sea = Mul_Add_sea.predict(test)
rmse_Mult_add_sea = np.sqrt(np.mean((np.array(test['Close'])-np.array(np.exp(pred_Mult_add_sea)))**2))
rmse_Mult_add_sea 
```

## Differencing

```python
plt.figure(figsize=(11,6))
fig = plt.figure(1)
ts_log_diff = ts_log - ts_log.shift(periods=1) # I will shift the time series by 1 and subtract from the original TS.
plt.plot(ts_log_diff)

#Determing rolling statistics
rolLmean = ts_log_diff.rolling(12).mean()
rolLstd = ts_log_diff.rolling(12).std()

#Plot rolling statistics:
orig = plt.plot(ts_log_diff, color='blue',label='Original')
mean = plt.plot(rolLmean, color='red', label='Rolling Mean')
std = plt.plot(rolLstd, color='black', label = 'Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```
